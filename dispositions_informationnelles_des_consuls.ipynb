```python
def uniques(df_col):
  return reduce(
      lambda x,y: x.union(y),
      filter(
          lambda x: x != '',
          map(set, df_col)
      )
  )
```

## Define Parsing Functions

```python
from functools import reduce

def parse_list_espace(x):
  l = []
  for e in x.split(';'):
    l.append(e.lower().strip().title())
  return l
```

```python
def find_n_elements_around_word(word_list, target_word, n):
    if isinstance(word_list, str):
      word_list = word_list.split(' ')

    occurrences = set()
    for i, word in enumerate(word_list):
        if word == target_word:
            start_index = max(0, i - n)
            end_index = min(len(word_list), i + n + 1)
            before = " ".join(word_list[start_index:i])
            after = " ".join(word_list[min(i+1,end_index):end_index])
            occurrences.add(before)
            occurrences.add(after)

    return occurrences


def get_spaces_from(df_col, not_spaces=set()):
  """
  Extracts "spaces" from a DataFrame column containing sentences.

  Args:
      df_col (pandas.Series): A pandas Series representing the DataFrame column containing sentences.
      not_spaces (dict, optional): A dictionary containing words that should not be considered as spaces.
          If not provided, a default set of words will be used. Default is None.

  Returns:
      pandas.Series: A pandas Series containing sets of extracted "spaces" for each sentence in the input column.

  Notes:
      - A "space" is defined as a titlecased word (starting with an uppercase letter) that is not present in the
        `not_spaces` dictionary.
      - The function splits each sentence into individual words, strips punctuation marks from the words, and checks
        if each titlecased word is a valid "space".
      - The returned Series contains a set of extracted "spaces" for each sentence. Duplicate spaces within a sentence
        are eliminated.

  Example:
      >>> df = pd.DataFrame({'sentences': [
        "Mr de Langallerie est soupçonné d'être un espion.",
        "Le roy d'angleterre est attendu à Hannovre",
        "Le roy de prusse est allé faire un tour au païs de cleue."
      ]})
      >>> spaces = get_spaces_from(df['sentences'])
      >>> print(spaces)
      0              {langallerie}
      1                 {Hannovre}
      2                         {}
      Name: sentences, dtype: object
  """
  not_spaces.update({
      'La', 'Le', 'Les',
      'Je', 'On', 'Il',
      'Une', 'Un',
      'De', 'Du', 'Des',
      'Ce', 'Cette',
      'Deux',
      'Enfant'
  })
  not_spaces.update({e.title() for e in MOTS_COURANTS})
  not_spaces.update({e.title() for e in MOTS_TROUVÉS})

  def extract_spaces(sentence):
    extracted_spaces = {}
    words = sentence.split(' ')
    rest_of_the_sentence = words[1:]
    # Iterate over all but the first word in the sentence
    for word in rest_of_the_sentence:
      word = word.strip(' .,:;?!"%&-_')
      if word.istitle() and word not in not_spaces:
        extracted_spaces[word] = find_n_elements_around_word(
            sentence,
            word,
            NUMBER_OF_WORDS_IN_CONTEXT
        )
    return extracted_spaces

  return df_col.apply(extract_spaces)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def create_bar_plot(
    data: pd.DataFrame,
    column_x: str,
    column_y: str,
    label_x: str,
    label_y: str,
    title: str,
    color: str = 'steelblue',
    bar_width: float = 0.8,
    handle_missing: str = None  # 'ignore', 'zero', 'interpolate'
):
    if handle_missing is None:
        pass
    elif handle_missing == 'ignore':
        data = data.dropna()
    elif handle_missing == 'zero':
        data[column_y].fillna(0, inplace=True)
    elif handle_missing == 'interpolate':
        data[column_y].interpolate(inplace=True)

    # Create the bar plot
    plt.figure(figsize=(14, 6))
    bars = plt.bar(data[column_x], data[column_y], width=bar_width, color=color)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)

    # Add labels to the bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom')

    # Rotate x-axis labels
    plt.xticks(data[column_x], rotation=45, ha='right')

    # Adjust spacing to prevent label overlapping
    plt.subplots_adjust(bottom=0.3)

    # Display the plot
    plt.tight_layout()
    plt.show()
```

```python
import unicodedata
from collections import Counter


def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return u"".join([c for c in nfkd_form if not unicodedata.combining(c)])

normalized_spaces = {
    remove_accents(e)
    for e in known_espaces
    if e.istitle() and e != ''
}

information_espaces = df['résumé'].drop_duplicates().apply(lambda x: [
    y.lower().capitalize()
    for y in x.replace('.', ' ').replace(',', ' ').split(' ')
    if remove_accents(y.lower().capitalize()) in normalized_spaces
])


def word_occurrence_dataframe(column, filtering):
    # Concatenate all lists of words into a single list
    all_words = filter(
        filtering,
        [word for sublist in column for word in set(sublist)]
    )

    # Count word occurrences using Counter
    word_counts = Counter(all_words)

    # Create a new DataFrame from word counts
    word_occurrence_df = pd.DataFrame(word_counts.items(), columns=['Espaces', 'Occurrences'])

    # Sort the DataFrame by word occurrences
    word_occurrence_df = word_occurrence_df.sort_values(by='Occurrences', ascending=False)

    return word_occurrence_df

def is_related_to_allemagne(x):
  child = x
  for _ in range(10):
    if child == 'Allemagne':
      return True
    child = parents.get(child)
```
